```python
data.loc[data["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA"]
```

```python
# combine all filters together
data.loc[(data["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA") & \
         (data["Event distance/length"].isin(["50km","50mi"])) & (data["Year of event"] == 2020)]
```

```python
df = data.loc[(data["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA") & \
         (data["Event distance/length"].isin(["50km","50mi"])) & (data["Year of event"] == 2020)].copy()
```

```python
df.head(10)
```

```python
df["Event name"] = df["Event name"].str.split("(").str.get(0)
```

```python
df["Athlete age"] = df["Year of event"] - df["Athlete year of birth"]
```

```python
# remove h from Athlete performance
df["Athlete performance"] = df["Athlete performance"].str.split(" ").str.get(0)
```

```python
# drop some columns
df = df.drop(["Athlete club"	,"Athlete country","Athlete year of birth" ,"Athlete age category"], axis = 1)
```

```python
# clean up null values
df.isna().sum()
```

```python
df[df["Athlete age"].isna() == 1]
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df[df.duplicated() == 1]
```

```python
df.reset_index(drop = True)
```

```python
# fix types
df.dtypes
```

```python
df["Athlete age"] = df["Athlete age"].astype(int)
```

```python
df["Athlete average speed"] = df["Athlete average speed"].astype(float)
```

```python
df = df.rename(columns = {"Year of event": "year",
                    "Event dates": "race_day",
                    "Event name": "race_name",
                    "Event distance/length":"race_length",
                    "Event number of finishers": "race_number_of_finishers",
                    "Athlete performance":"athlete_performance",
                    "Athlete gender":"athlete_gender",
                    "Athlete average speed":"athlete_average_speed",
                    "Athlete ID": "athlete_id",
                    "Athlete age": "athlete_age"})
```

```python
df1 = df[["race_day","race_name","race_length","race_number_of_finishers","athlete_id","athlete_gender","athlete_age","athlete_performance","athlete_average_speed"]]
```

```python
# cahrts and graphs
```

```python
sns.histplot(df1["race_length"])
```

```python
sns.histplot(df1, x= "race_length", hue = "athlete_gender")
```

```python
sns.lmplot(df1, x = "athlete_age", y ="athlete_average_speed", hue = "athlete_gender")
```

```python
# difference in speed for 50km, 50mile of female and male 
df1.groupby(["race_length","athlete_gender"])["athlete_average_speed"].mean()
```
